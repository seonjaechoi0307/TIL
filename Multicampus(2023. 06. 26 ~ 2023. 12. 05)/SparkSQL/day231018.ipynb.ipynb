[공지 및 조언]
- 최종 프로젝트에서 핵심은 수집된 데이터를 빅쿼리 등 DBMS로 저장한 후, DB에서 분석할 데이터를 추출하는 것이 기본 출발점입니다.

- IT/빅데이터 기술이 메인인 회사에서 데이터는 모두 빅쿼리와 같은 DBMS에 있기 때문에, 데이터 수집부터 쿼리를 통해 데이터를 가져오게 됩니다. 이 부분이 포트폴리오 PPT에 잘 녹여 있어야 합니다.
따라서, 프로젝트를 위한 데이터 수집 시, 데이터 적재는 모두 빅쿼리로 해주시기를 바랍니다.

- 그 외 MySQL, PostgreSQL 제품을 사용하는 것도 자유이지만, 인스턴스 생성 시 부터 비용이 부과되는 점만 기억하셔서 향후에 어떤 DB를 사용할지는 팀에서 최종 결정하시면 됩니다.

```python
# Streamlit 앱을 시작합니다.
import streamlit as st

# 데이터를 다루기 위한 pandas 라이브러리를 임포트합니다.
import pandas as pd

# Google Cloud 서비스와 통합하기 위한 라이브러리를 임포트합니다.
from google.auth import compute_engine
from google.cloud import bigquery

# Google Cloud 계정 인증 정보를 설정합니다.
credentials = compute_engine.Credentials(
    service_account_email='your_email_key')

# BigQuery 클라이언트를 설정합니다.
client = bigquery.Client(
    project='your_project_id',
    credentials=credentials)

# 쿼리 결과를 캐시하고 10분 동안 재실행하지 않도록 설정합니다.
@st.cache_data(ttl=600)
def run_query():
    # BigQuery에서 데이터를 쿼리하는 SQL 문을 정의합니다.
    sql = """
      SELECT * FROM `your_DB.Skemas.table` LIMIT 1000;
    """

    # BigQuery 프로젝트 ID를 설정합니다.
    project_id = 'your_project_id'

    # BigQuery 쿼리를 실행하고 결과를 pandas 데이터프레임으로 변환합니다.
    df = pd.read_gbq(sql, project_id=project_id, dialect='standard')

    # 결과를 Streamlit 앱에 출력합니다.
    st.write(df)

    # 제목을 추가합니다.
    st.title("The query data 가져오기")

# Streamlit 앱의 주요 함수를 정의합니다.
def main():
    # 앱의 제목을 설정합니다.
    st.title("Streamlit Iris Data Visualization")

    # 데이터 쿼리를 실행하는 함수를 호출합니다.
    run_query()

    # 앱 내용을 추가합니다.
    st.write("""
    ## Explore the Iris dataset
    Use the controls below to explore different visualizations of the Iris dataset.
    """)

# 스크립트가 직접 실행될 때 main 함수를 호출합니다.
if __name__ == '__main__':
    main()
```

```python
import streamlit as st
import pandas as pd
from google.auth import compute_engine
from google.cloud import bigquery
credentials = compute_engine.Credentials(
    service_account_email='your_email_key')

client = bigquery.Client(
    project='your_project_id',
    credentials=credentials)

# Perform query.
# Uses st.cache_data to only rerun when the query changes or after 10 min.
@st.cache_data(ttl=600)
def run_query():
    sql = """
      SELECT * FROM `your_DB.Skemas.table` LIMIT 1000
    """

    project_id = 'your_project_id'
    df = pd.read_gbq(sql, project_id = project_id, dialect='standard')
    st.write(df)

    st.title("The query data 가져오기")


def main():

    st.title("Streamlit Iris Data Visualization")
    run_query()
    st.write("""
    ## Explore the Iris dataset
    Use the controls below to explore different visualizations of the Iris dataset.
    """)



if __name__ == '__main__':
    main()
```
