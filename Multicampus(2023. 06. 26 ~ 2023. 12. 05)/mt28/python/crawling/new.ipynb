```python
%%timeit

import requests
import lxml
from bs4 import BeautifulSoup
import pandas as pd
url = "http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key=1308370912&type=xml&iStartUnitCode=101&iEndUnitCode=103"
req = requests.get(url)
# print(req.text)
soup = BeautifulSoup(req.text, "lxml") # XML 반환값이라서 lxml을 사용한 것
# print(soup.prettify())
## 미션 XML 결괏값으로 데이터프레임으로 변환 (11시 10분까지)
stddate_list = soup.find_all("stddate")
timeAvg_list = soup.find_all("timeavg")
stddate = []
avgTime = []
for date, timeAvg in zip(stddate_list, timeAvg_list):
    # print(date.get_text(), timeAvg.get_text())
    stddate.append(date.get_text())
    avgTime.append(timeAvg.get_text())
df = pd.DataFrame({"date":stddate, "avgTime" : avgTime})
df
# timeAvg_list
```

```python
import requests
import json
import pandas as pd

# 변수설정
# {"tbLnOpendataRtmsV": { "list_total_count": 1862 = max_num
SERVICE_KEY = '564b7852646a686a34336f4f6c5571'
max_num = data_json['tbLnOpendataRtmsV']['list_total_count'] # 최종페이지
df_list = [] # 데이터 프레임 리스트를 담을 변수

for start_num in range(1, max_num, 1000):
    end_num = start_num + 999 # 갖고올 페이지 범위 시작 페이지 + 999 = 최대 1000 page만 가져올 수 있도록
    if end_num > max_num: # end_num이 max보다 클 시 end_num 숫자를 max_num으로 변환
        end_num = max_num
    url = f'http://openapi.seoul.go.kr:8088/{SERVICE_KEY}/json/tbLnOpendataRtmsV/{start_num}/{end_num}/2023/11560'
    req = requests.get(url)
    content = req.json() # req.json() 정보 content 변수로 저장
    df = pd.DataFrame(content['tbLnOpendataRtmsV']['row']) # content 변수 저장값 데이터 프레임화 시켜서 df 변수에 저장
    df_list.append(df) # df_list 에 append 로 df 값들 저장

result_df = pd.concat(df_list, ignore_index=True) # pd.concat 으로 데이터 프레임 합치기
result_df # 합친 데이터 프레임 출력
result_df.to_csv("ATP_data.csv", index=False, encoding='utf-8-sig')

# 과제
# 시작 인덱스 1, 끝 인덱스는 12000 반복문 통해서
# 시작 인덱스와 끝 인덱스를 조정하면, 원하는 데이터가 출력
# 시작 1 ~ 1000 -> df1
# 시작 1001 ~ 2000 -> df2
# 시작 2001 ~ 3000 -> df3

    
```
