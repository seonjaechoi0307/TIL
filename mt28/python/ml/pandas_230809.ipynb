## 행 추가

```python
df = pd.DataFrame({
    "이름": ["Evan", "Sara", "Chris", "Messi", "Woods", "Jordan"], 
    "국어": [70, 90, None, None, 100, 85], 
    "영어": [65, 100, None, 90, 70, 80]
})

df
```

```python
# 행 추가
df.loc[6, "이름":"영어"] = ["kim", 50, 60]
df
```

```python
%%timeit
# 국어 점수가 90점 이상이면 "A"
# 국어 점수가 80점 ~ 89점 "B"
# 국어 점수가 70점 ~ 79점 "C"
# 나머지는 F

# 성적기반 등급분류 함수
def get_grade(score):
    if score >= 90:
        return 'A'
    elif 80 <= score < 90: 
        return 'B'
    elif 70 <= score < 80: 
        return 'C'
    else:
        return 'F'
        
# 국어 점수기반으로 함수를 사용해 국어점수 등급분류
df['등급'] = df['국어'].apply(get_grade)
df
```

 - numpy.where, 조건식 한번만 적용
 if 조건식, 참일 때, 거짓일 때

```python
# np.where(조건식, 참, np.where(조건식, 참, np.where(조건식, 참)))
# np.select(다중조건식, 다중결괏값, default = else 조건)
```

```python
%%timeit
condition_list = [(df["국어"] >= 90), (df['국어'] >= 80) & (df['국어'] < 90), (df['국어'] >= 70) & (df['국어'] < 80)]
choice_list = ["A", "B", "C"]
df['점수'] = np.select(condition_list, choice_list, default = "F")
df
```

## 코로나 관련 변수만 추출

```python
totvars_misscnt = covidtotals[totvars].isnull().sum(axis=1)
totvars_misscnt.value_counts()
ms_country = covidtotals.loc[totvars_misscnt > 1, ['location'] + totvars]
ms_country.T
```

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly['total_cases_q'] = pd.qcut(covidtotalsonly['total_cases'],
                                          labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q = 5, precision=0)

covidtotalsonly['total_deaths_q'] = pd.qcut(covidtotalsonly['total_deaths'],
                                          labels = ['매우낮음', '낮음', '중간', '높음', '매우높음'],
                                          q = 5, precision=0)
covidtotalsonly
```

```python
pd.crosstab(covidtotalsonly['total_cases_q'], covidtotalsonly['total_deaths_q'])
```
